```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Định nghĩa không gian nền x (Số lượng sinh viên từ 0 đến 200)
x = np.linspace(0, 200, 500)

# 2. Định nghĩa hàm thuộc cho các tập mờ cơ sở
# Tập A: Phòng rộng (giả sử rộng bắt đầu từ 60 chỗ và đạt mức 1 tại 140 chỗ)
mu_A = np.clip((x - 60) / (140 - 60), 0, 1)

# Tập B: Chứa khoảng 100 sinh viên (dùng hàm Gaussian centered at 100)
mu_B = np.exp(-0.5 * ((x - 100) / 20)**2)

# 3. Tính toán các tập mờ theo yêu cầu bài tập
mu_D = np.minimum(mu_A, mu_B)  # Giao (A ∩ B)
mu_E = np.maximum(mu_A, mu_B)  # Hợp (A ∪ B)
mu_F = 1 - mu_A                # Phủ định (¬A)

# 4. Vẽ biểu đồ
plt.figure(figsize=(12, 8))

# Vẽ A và B (nền)
plt.plot(x, mu_A, 'b--', label='A: Phòng rộng', alpha=0.5)
plt.plot(x, mu_B, 'g--', label='B: Chứa khoảng 100 SV', alpha=0.5)

# Vẽ các kết quả
plt.plot(x, mu_D, 'red', linewidth=2, label='D: Rộng VÀ Chứa ~100 (A ∩ B)')
plt.plot(x, mu_E, 'purple', linewidth=2, label='E: Rộng HOẶC Chứa ~100 (A ∪ B)')
plt.plot(x, mu_F, 'orange', linewidth=2, label='F: Không rộng (¬A)')

# Định dạng biểu đồ
plt.title('Biểu diễn các tập mờ D, E, F', fontsize=14)
plt.xlabel('Sức chứa (Số lượng sinh viên)')
plt.ylabel('Mức độ thuộc $\mu(x)$')
plt.legend()
plt.grid(True, linestyle=':', alpha=0.6)
plt.axhline(0, color='black', lw=1)
plt.axvline(0, color='black', lw=1)

plt.show()
```
